```python
file3 = file3.rename(columns={"State": "ST", "Gender": "GENDER"})
```

```python
#concatenate files
data_set = pd.concat([file1, file2, file3], axis=0)
data_set
```

```python
#identify categorical columns
data_set.select_dtypes(include=object)
```

<h2>Data Cleaning Operations</h2>

```python
#drop 'education' and 'number_of_open_complaints' columns
columns_drop = ['education', 'number_of_open_complaints']
```

```python
data_set = data_set.drop(columns_drop, axis = 1)
```

```python
data_set.info()
```

```python
nulls_df = pd.DataFrame(round(data_set.isna().sum()/len(data_set)*100,2))
```

```python
#correct 'customer_lifetime_value' from percentage to numerical value
clv_num = data_set['customer_lifetime_value']
clv_num
```

```python
type(clv_num)
```

```python
clv_num = data_set['customer_lifetime_value'].str.replace('%','')
```

```python
clv_num
```

```python
def function(a):
    if a != 'NaN':
        a = str(a).replace('%','')
    return float(a)

clv_num = data_set['customer_lifetime_value'].apply(function)
print(clv_num)
```

```python
#replacing NaN with numerical 0 value
clv_num = data_set['customer_lifetime_value'].fillna(value = 0)
clv_num
```

```python
data_set['customer_lifetime_value'] = clv_num
data_set
```

```python
#check for duplicate rows
data_set.drop_duplicates()
```

```python
#filter out any 'income' less than 0
df_filtered = data_set[data_set['income'] > 0]
```

```python
data_set = df_filtered
```
